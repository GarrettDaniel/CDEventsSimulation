```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from copy import deepcopy

pd.set_option('display.max_columns', None)
```

```python
events = pd.read_csv("../simulated_data/simulated_processed_events.csv")
events.head()
```

```python
label = 'run_outcome'
```

```python
relevant_cols = [
    'event_id', 'context_id', 'context_source',
    'context_version', 'context_type', 'context_timestamp',
    'subject_type', 'run_outcome', 'content_task', 'run_pipelineName'
]

events_rel = deepcopy(events[relevant_cols])
events_rel.head()
```

```python
events_rel[label].value_counts(dropna=False)
```

## Prepare Data for ML Model

### Step 1: Feature Engineering

### Step 1a: Create Features

Features to create
- Time since pipeline started
- Time in queued state (for pipelineRun events only)
- user
- environment

```python
## user and environment
events_rel['context_source'].str.split("/", expand=True).head()
```

```python
## event_state
events_rel['context_type'].str.split(".", expand=True).head()
```

```python
## event_state, environment, user

context_type_expanded = events_rel['context_type'].str.split(".", expand=True)
environment = context_type_expanded[0]
event_state = context_type_expanded[3]

context_source_expanded = events_rel['context_source'].str.split("/", expand=True)
user = context_source_expanded[2]

events_rel['event_state'] = event_state
events_rel['environment'] = environment
events_rel['user'] = user

events_rel.drop(columns=['context_source', 'context_type'], inplace=True)

events_rel.head()
```

```python
events_rel['context_id'].value_counts()
```

Create time-based columns (`time_since_started`, `time_queued`)

```python
from datetime import datetime, timedelta

## Convert time column to datetime
events_rel['context_timestamp'] = pd.to_datetime(events_rel['context_timestamp'])
```

```python
time_map = {}

queued = None
time_queued = None
time_since_start = None

started = time_example_df[time_example_df['event_state'] == "started"]['context_timestamp'].values[0]
finished = time_example_df[time_example_df['event_state'] == "finished"]['context_timestamp'].values[0]

if "queued" in time_example_df['event_state'].unique():
    queued = time_example_df[time_example_df['event_state'] == "queued"]['context_timestamp'].values[0]
    time_since_start = finished - queued
    time_queued = started - queued
    
else:
    time_since_start = finished - started
    
time_map['time_queued'] = time_queued.astype('timedelta64[s]').item().total_seconds()
time_map['time_since_start'] = time_since_start.astype('timedelta64[s]').item().total_seconds()

time_map
```

Now let's do it for all event contexts

```python
time_map = {}

for context_id in events_rel['context_id'].unique():
    time_map[context_id] = {}
    time_df = events_rel[events_rel['context_id'] == context_id]

    queued = None
    time_queued = None
    time_since_start = None

    started = time_df[time_df['event_state'] == "started"]['context_timestamp'].values[0]
    finished = time_df[time_df['event_state'] == "finished"]['context_timestamp'].values[0]

    if "queued" in time_df['event_state'].unique():
        queued = time_df[time_df['event_state'] == "queued"]['context_timestamp'].values[0]
        
        time_since_start = finished - queued
        time_queued = started - queued
        
        time_map[context_id]['time_queued'] = time_queued.astype('timedelta64[s]').item().total_seconds()

    else:
        time_since_start = finished - started
        time_map[context_id]['time_queued'] = None

    time_map[context_id]['time_since_start'] = time_since_start.astype('timedelta64[s]').item().total_seconds()

time_map
```

```python
time_queued_and_since_start = pd.DataFrame(time_map).T
time_queued_and_since_start.head(10)
```

```python
events_rel_finished = deepcopy(events_rel[events_rel[label].notnull()])
events_rel_finished_times = events_rel_finished.join(time_queued_and_since_start, on='context_id')
events_rel_finished_times['time_queued'] = np.where(events_rel_finished_times['time_queued'].isnull(), 0, events_rel_finished_times['time_queued'])
```

```python
events_rel_finished_times.head(10)
```

### Step 1b: One-hot Encoding

Features to one-hot encode
- environment
- user
- subject_type
- context_version
- content_task
- run_pipelineName

```python
to_encode = ['environment', 'user', 'subject_type', 'context_version', 'content_task', 'run_pipelineName']

encoded = pd.get_dummies(data=events_rel_finished_times, columns=to_encode)

events_rel_finished_times.drop(columns=to_encode, inplace=True)
events_rel_finished_times[encoded.columns] = encoded

events_rel_finished_times.head(10)
```

## Step 2: Train

### Step 2a: Split into train and test sets

```python
from sklearn.model_selection import train_test_split

cols_to_drop = [
    'event_id',
    'context_id',
    'context_timestamp',
    'event_state'
]

event_rel_train_test = deepcopy(events_rel_finished_times.drop(columns=cols_to_drop))
```

```python
event_rel_train_test.head(10)
```

```python
X = deepcopy(event_rel_train_test)
y = X.pop(label)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

### Step 2b: Preprocessing (SMOTE and Scaling)

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

oversample = SMOTE()
scaler = StandardScaler()

numeric_columns = ['time_queued', 'time_since_start']
```

```python
y_train.value_counts()
```

```python
X_train[numeric_columns] = scaler.fit_transform(X_train[numeric_columns])
```

```python
X_test[numeric_columns] = scaler.fit_transform(X_test[numeric_columns])
```

```python
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```python
y_train.value_counts()
```

```python
X_train.head()
```

### Step 2c: Train and Test

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier, VotingClassifier
```

```python
estimators = []

gb = GradientBoostingClassifier()
rf = RandomForestClassifier()
ab = AdaBoostClassifier()

estimators.append(('GradientBoost', gb))
estimators.append(('RandomForest', rf))
estimators.append(('AdaBoost', ab))

ensemble = VotingClassifier(estimators=estimators, voting='hard')

ensemble.fit_transform(X_train, y_train)
```

### Step 3: Predict and Evaluate

Especially since this is randomly simulated data with no patterns intentionally created, we expect that the accuracy for these models will be low.  The idea here is to show a proof-of-concept pipeline, not to achieve highly accurate predictions on simulated data without patterns.

```python
y_pred = ensemble.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

This project could be further improved with additional methods such as cross-validation, using real pipeline data instead of simulated data, better feature selection, and hyperparameter tuning.  The intent was to show an example of machine learning work that could be done on this kind of data.
